```python
num_rows = len(df) # should be 2729, 行数

for row in range(num_rows):
# 修改一行数据的顺序
    temp = []
    for i in range(0, 16+1, 8):
        # 将相关的24个数据分成三段（每8个属于一辆车），并变成一个长度为3的复合list以便排序
        section = list(df.iloc[row][74:98])[i:i+8]
        temp.append(section)

    # 排序
    temp = sorted(temp)
    # 特殊情况
    if temp[1][0] == 3 and np.isnan(temp[2][0]):
        temp[1], temp[2] = temp[2], temp[1]
    # flat the list （展平）
    temp = [item for sublist in temp for item in sublist]

    # 替换进去
    df.iloc[row, 74:98] = temp


dict(df.iloc[130])
```

```python
"""需求：没有转向灯为0，左为1，右为2，双闪为3
左右灯：L为0x0020, R 0x0040
双闪：0x4000
"""

def str_strip(x):
    # Otherwise it's a str in 0x[][][][] format
    if type(x)==str and len(x)>=2 and x[-2] == str(2):
        return 1
    elif type(x)==str and len(x)>=2 and x[-2] == str(4):
        return 2
    elif type(x)==str and len(x)>=4 and x[-4] == str(4):
        return 3
    else:
        # we don't care about other lightmasks, like brake light
        return 0
    
for i in range(1,6):
    df["lightMask.{}".format(i)] = df["lightMask.{}".format(i)].apply(str_strip)
    print(pd.value_counts(df["lightMask.{}".format(i)]))

df["lightMask"] = df["lightMask"].apply(str_strip)
dict(pd.value_counts(df["lightMask"]))
```
