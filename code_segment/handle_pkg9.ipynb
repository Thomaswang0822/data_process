```python
"""
This file should handle the raw csv of pkg7 produced by VTD
"""
import pandas as pd
import numpy as np
import csv
```

```python
"""
All constants defined here, and we only need to modify here in the future, hopefully.
"""


# A
DATA_DIR = "../raw_data/0810/0810_2_24/"
SCENE = "2_24"
PKG_ID = '9'
FILE_PATH = DATA_DIR + SCENE + "pkg{}.csv".format(PKG_ID)

OUT_DIR = "../processed/processed_0810/"
OUTPUT_PATH = OUT_DIR+SCENE+"pkg{}_del.csv".format(PKG_ID)

# B
"""
# We want objectId, X, Y, yaw_angle(h after X), Vx, Vy, Ax, Ay as well as dimXYZ, offXYZ for all cars
# col index see below (*range(7,13) is 7~12, dimXYZ and offXYZ)
"""
index_to_keep = [2, 4, *range(7,13), 13, 14, 16, 22, 23, 31, 32] # remainder of index//PATTERN
PATTERN = 39  # number of entries of one road line, which form a pattern
col_of_one = ["objectId", 'obj_type','dimX','dimY','dimZ','offX','offY','offZ', "X", "Y", "yaw_angle", "Vx","Vy", "Ax", "Ay"]


PLACE_HOLDER = -99.99
```

```python
with open(FILE_PATH, 'r', encoding="utf-8") as csvfile:
    reader = csv.reader(csvfile)
    MAX_LEN = max(len(_) for _ in reader)
    
MAX_LEN
```

```python
original_col_names = ["temp.{}".format(i) for i in range(MAX_LEN)]
df9 = pd.read_csv(FILE_PATH, names=original_col_names, low_memory=False)        

df9.drop(index=0, axis=0, inplace=True)
df9.index -= 1 
df9.columns = [name.strip() for name in df9.columns]
      
df9.tail(10)
```

```python
rows = df9.shape[0]
num_del = rows%100
df9.drop(labels=range(rows-num_del, rows), axis=0, inplace=True)
df9.shape
```

```python
"""
Other constants which must be defined after reading in df.


Note: when determing number of objs, we use (#cols-2)/PATTERN.
    But quite often, the raw data has an empty column in the end,
    with pandas automatically giving a name "Unnamed.0"
"""
# C
while (len(df9.columns)-2)%PATTERN != 0:
    # del that col
    print("Found an empty col at the end")
    df9.drop(df9.columns[-1], axis=1, inplace=True)
    
NUM_OBJS = (len(df9.columns)-2)//PATTERN
BEGIN_IDX = NUM_OBJS # current df has 10 roadlines, 0-9
TARGET = 10 # we want 14
END_IDX = 10

# Check if it is a whole number
NUM_OBJS
```

```python
"""### Keep those cols we want and check ###"""

# a list of column names to delete
cols_to_del = [name for i, name in enumerate(df9.columns) if i%PATTERN not in index_to_keep and i!=1]
df9 = df9.drop(columns=cols_to_del)
df9.head(10)
```

```python
"""### do some simple pre-process and check ###"""
df9.fillna(PLACE_HOLDER, inplace=True)
df9.rename(columns={"temp.1":"simFrame"}, inplace=True)
df9["simFrame"] = df9["simFrame"].astype(np.int64)
df9.set_index("simFrame", inplace=True)
df9 = df9[~df9.index.duplicated(keep='first')]

"""
# rename col names of first chunk: id to id.0
rename_dict = {name:name+'.0' for name in col_of_one}
df9.rename(columns=rename_dict, inplace=True)

# rename type.x to obj_type.x to distinguish (pkg 7 roadmark also has type)
type_dict = {"type.{}".format(i): "obj_type.{}".format(i) for i in range(NUM_OBJS)}
df9.rename(columns=type_dict, inplace=True)
"""

new_names = ["{}.{}".format(name,i) for i in range(NUM_OBJS) for name in col_of_one]
df9.columns = new_names

for i in range(NUM_OBJS):
    df9["{}.{}".format(col_of_one[0], i)] = df9["{}.{}".format(col_of_one[0], i)].apply(lambda x: int(x.strip()) if type(x)==str else x)
    print( df9["{}.{}".format(col_of_one[0], i)].value_counts())
    
# append empty colunms to match designated number of lines reserved: 14
"""
Idea explained in pkg 7 file
"""
for i in range(BEGIN_IDX, END_IDX):
    for name in col_of_one:
        df9[name+'.'+str(i)] = PLACE_HOLDER
        
        
NUM_OBJS = END_IDX

print(df9.shape)
df9.head(10)
```

```python
"""
MOST important task: Standardization.

In many scenarios, not all objects will stay in the scene and recorded by the sensor, 
which results in mis-aligned chunks of cols for each object. For example
id0 id1 id2 id3
1   2   3   4
2   4   NaN NaN
............

TODO:
A sorting algorithm based on selection sort

Idea:
for each list(row):
    for each entry where objectId will be in, (a fixed jump of 14 here):
        group them and put into the place in the new list
        
    populate remaining entries, where the corresponding id is not in our data, with [NaN]*14
    new_list looks like [[data of id1], [data of id2], ......]
    flat the list
    list(row) = new_list
    
    
!!! Take quite a long time
"""
row_nums = df9.shape[0]    # 8336 this time
jump = len(col_of_one)   # 14

for row in range(row_nums):
    old_row = df9.iloc[row]
    new_row = [ [PLACE_HOLDER]*jump ]*NUM_OBJS


    for idx in range(0, 0 + NUM_OBJS*jump, jump):   # 0, 14, 28, ...
        objId = old_row[idx]
        if objId == PLACE_HOLDER or objId >= NUM_OBJS or objId <= 0:
            # Special case where id is placeholder -99.99 (also those trailing data)
            continue
        else:
            objId = int(objId)
            # ego car id=1 should go to objectId_0, id=5 should go to objectId=4 likewise.
            new_row[objId] = old_row[idx:idx+jump]
            
    # flat the list
    new_row = [item for sublist in new_row for item in sublist]
    df9.iloc[row] = new_row

              
df9.head(10) 
```

```python
for i in range(NUM_OBJS):
    print( df9["{}.{}".format(col_of_one[0], i)].value_counts())
```

```python
# do the check after the sort, there should be no output if it's correct
for i in range(END_IDX):
    assert len(df9["{}.{}".format(col_of_one[0], i)].value_counts()) <= 2, "objectId.{} col still has mixed id, need to further enlarge TARGET {}".format(i, TARGET)
```

```python
new_names = ["{}_{}".format(name,i) for i in range(NUM_OBJS) for name in col_of_one]
df9.columns = new_names 
df9.columns
```

```python
df9.to_csv(OUTPUT_PATH)
```
