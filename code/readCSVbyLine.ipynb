```python
with open(FILE_PATH, 'r', encoding="utf-8") as csvfile:
    reader = csv.reader(csvfile)
    MAX_LEN = max(len(_) for _ in reader)
    
MAX_LEN
```

```python
original_col_names = ["temp.{}".format(i) for i in range(MAX_LEN)]
df = pd.read_csv(FILE_PATH, names=original_col_names)
df
```

```python
original_col_names = ["temp.{}".format(i) for i in range(MAX_LEN)]
df = pd.DataFrame(columns=original_col_names)

with open(FILE_PATH, 'r', encoding="utf-8") as csvfile:
    reader = csv.reader(csvfile)
    for i, row in enumerate(reader):
        length = len(row)
        std_row = row + ['']*(MAX_LEN-length)
        df.loc[i] = std_row
        

df.drop(index=0, axis=0, inplace=True)
df.index -= 1        
      
df
```
